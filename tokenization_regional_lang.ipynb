Dealing with bengali characters can be a little tricky as
Grapheme cluster: the “thing your eyes see as one character” (e.g., ক + ি → কি). We must treat this as one unit.
Normalization (NFC): same letter can be stored in different internal forms. Normalize so equal things compare equal.
regex with \X: the pattern \X returns full grapheme clusters, not code points.

```python
#Now we need to find out the consicutively common elements of occurance by iterating through the text
import unicodedata
import regex
from collections import Counter
from itertools import pairwise
def count_consecutive_pairs_of_bengali_token(x):
    # Normalize so composed/combining forms are consistent
    text = unicodedata.normalize("NFC",x)

    # Extract user-perceived characters. \X = one grapheme cluster.
    graphemes = regex.findall(r'\X', text)
    # Encode each grapheme into its UTF-8 bytes → tuple of ints
    grapheme_bytes = [tuple(g.encode("utf-8")) for g in graphemes]
    
    consecutive_pairs = list(pairwise(grapheme_bytes))
    return Counter(consecutive_pairs)

def count_consecutive_pairs_of_bengali_text(x):
    # Normalize so composed/combining forms are consistent
    text = unicodedata.normalize("NFC",x)

    # Extract user-perceived characters. \X = one grapheme cluster.
    graphemes = regex.findall(r'\X', text)
    
    consecutive_pairs = list(pairwise(graphemes))
    return Counter(consecutive_pairs)
```

```python
def bengali_utf8_bpe(text, max_merges=10):
    # Normalize for consistent Unicode form
    text = unicodedata.normalize("NFC", text)

    # Split into graphemes
    graphemes = regex.findall(r'\X', text)

    # Encode each grapheme into its UTF‑8 bytes and flatten into a byte list
    sequence = []
    for g in graphemes:
        sequence.extend(list(g.encode("utf-8")))  # each byte 0–255

    print(f"Initial length: {len(sequence)} tokens (bytes)")
    print(f"First 40 tokens: {sequence[:40]}\n")

    next_token_id = 256  # IDs above 255 will be merged units
    merges = {}          # merge table for decoding

    # Perform merges
    for merge_num in range(1, max_merges + 1):
        # Count all consecutive byte pairs in the flat sequence
        pair_counts = Counter(pairwise(sequence))
        if not pair_counts:
            print("No more pairs to merge. Stopping.")
            break

        (most_common_pair, freq) = pair_counts.most_common(1)[0]
        if freq < 2:
            print("No pair repeats. Stopping.")
            break

        print(f"[Merge {merge_num}] Most common pair {most_common_pair} occurs {freq} times.")

        # Save merge in the table for decoding
        merges[next_token_id] = most_common_pair

        # Replace this pair in the sequence with the new token ID
        new_sequence = []
        i = 0
        while i < len(sequence):
            if i < len(sequence) - 1 and (sequence[i], sequence[i+1]) == most_common_pair:
                new_sequence.append(next_token_id)
                i += 2  # skip both elements of the merged pair
            else:
                new_sequence.append(sequence[i])
                i += 1

        sequence = new_sequence
        print(f"→ Created token ID {next_token_id} for {most_common_pair}")
        print(f"→ Sequence length after merge: {len(sequence)}\n")
        next_token_id += 1

    # Final stats
    print("=== Final Summary ===")
    print(f"Initial length (bytes): {len(graphemes)*3} approx")  # for Bengali chars
    print(f"Final compressed length: {len(sequence)} tokens")
    reduction = ((len(graphemes)*3 - len(sequence)) / (len(graphemes)*3)) * 100
    print(f"Total reduction: {reduction:.2f}%")

    return sequence, merges
```

```python

compressed_tokens, merge_table = bengali_utf8_bpe(ben_content_to_tokenize, max_merges=10)
```

```python
def bpe_decode(token_sequence, merges, errors="strict"):
    """
    Turn a compressed list of token IDs (from our UTF-8 ) Byte Pair Encoding
    back into the original human readable Bengali text.

    token_sequence : list of integers
        The compressed tokens from our merge step. 
        - If <= 255, it's a raw UTF-8 byte. 
        - If >= 256, it's a merged token we must expand.

    merges : dict
        The merge table from encoding: 
        maps each merged token ID to the two IDs it came from.
    """

    def expand_token(token_id):
        """
        Take a token ID and turn it into the bytes that it represents.
        - If it's a byte (0-255), just return it in a list.
        - If it's a merged token (>=256), look up what two tokens it was made from,
          then expand each of those tokens in turn (recursively).
        """
        if token_id < 256:
            # it's already a raw byte
            return [token_id]
        else:
            # Look up the original two tokens this ID was merged from
            left, right = merges[token_id]

            # Recursively expand both sides to get their bytes
            return expand_token(left) + expand_token(right)

    # Expand every token in the compressed sequence back into raw bytes
    byte_ids = []
    for tid in token_sequence:
        byte_ids.extend(expand_token(tid))

    # Turn the list of byte integers into a bytes object
    raw_bytes = bytes(byte_ids)

    # Decode from UTF‑8 to get our original text
    # - Using errors="strict" will throw an error if we messed up merges
    # - Using errors="replace" will swap invalid sequences with '�'
    return raw_bytes.decode("utf-8", errors=errors)
```

```python
original_text = bpe_decode(compressed_tokens, merge_table, errors="strict")
print(original_text)
```
